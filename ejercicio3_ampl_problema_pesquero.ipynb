```python
ampl = ampl_notebook(modules=["highs"], license_uuid="default")
```

```python
model='''

#definicion de variables

var viajes_pacifico >= 0 integer;
var viajes_atlantico >= 0 integer;

# definicion de funcion objetivo

minimize costo: 13000 * viajes_pacifico  + 10000 * viajes_atlantico;

# definicion de restricciones

s.t. big_eye: 150 * viajes_pacifico + 100 * viajes_atlantico >= 1500;
s.t. yellowfin: 100 * viajes_pacifico + 200 * viajes_atlantico >= 600;
s.t. skipjack: 50 * viajes_pacifico + 100 * viajes_atlantico >= 700;

'''
```

```python
ampl.eval(model)
ampl.option["solver"]='highs'
ampl.solve()
```

```python
print("Resultados de los viajes: ")

cant_viajes_pacifico = ampl.get_variable('viajes_pacifico').value()
cant_viajes_atlantico = ampl.get_variable('viajes_atlantico').value()
profit = ampl.get_objective('profit').value()

print(("viajes hacia el pacifico: "), cant_viajes_pacifico)
print(("viajes hacia el atlantico: "), cant_viajes_atlantico)
print(("costo: "), costo)
```
