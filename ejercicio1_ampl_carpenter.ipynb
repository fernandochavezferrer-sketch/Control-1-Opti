```python
ampl = ampl_notebook(modules=["highs"], license_uuid="default")
```

```python
model='''

#definicion de variables

var mecedoras >= 0 integer;
var taburete >= 0 integer;
var mesa >= 0 integer;

# definicion de funcion objetivo

maximize profit: 70 * mecedoras + 15 * taburete + 45 * mesa;

# definicion de restricciones

s.t. peq_req: 18 * mecedoras + 2 * taburete + 4 * mesa <= 54;
s.t. med_req: 9 * mecedoras + 2 * taburete + 4 * mesa <= 30;
s.t. gra_req: 2 * mecedoras + 1 * taburete <= 30;

'''
```

```python
ampl.eval(model)
ampl.option["solver"]='highs'
ampl.solve()
```

```python
print("Resultados de la produccion: ")

cant_mecedoras = ampl.get_variable('mecedoras').value()
cant_taburete = ampl.get_variable('taburete').value()
cant_mesa = ampl.get_variable("mesa").value()
profit = ampl.get_objective('profit').value()

print(("mecedoras a fabricar: "), cant_mecedoras)
print(("taburetes a fabricar: "), cant_taburete)
print(("mesas a fabricar: "), cant_mesa)
print(("Beneficio: "), profit)
```
