```python
ampl = ampl_notebook(modules=["highs"], license_uuid="default")
```

```python
model='''

#definicion de variables

var camion_A >= 0 integer;
var camion_B >= 0 integer;

# definicion de funcion objetivo

minimize costo: 30 * camion_A  + 40 * camion_B;

# definicion de restricciones

s.t. prod_refrigerado: 20 * camion_A + 30 * camion_B >= 900;
s.t. prod_no_refrigerado: 40 * camion_A + 30 * camion_B >= 1200;


'''
```

```python
ampl.eval(model)
ampl.option["solver"]='highs'
ampl.solve()
```

```python
print("camiones necesarios: ")

cant_camiones_A = ampl.get_variable('camion_A').value()
cant_camiones_B = ampl.get_variable('camion_B').value()
costo = ampl.get_objective('costo').value()

print(("camiones A: "), cant_camiones_A)
print(("camiones B: "), cant_camiones_B)
print(("costo: "), costo)
```
