```python
ampl = ampl_notebook(modules=["highs"], license_uuid="default")
```

```python
model='''

#definicion de variables

var televisor >= 0 integer;
var equipo_sonido >= 0 integer;

# definicion de funcion objetivo

maximize profit: 200 * televisor + 120 * equipo_sonido;

# definicion de restricciones

s.t. req_semanal: televisor + equipo_sonido >= 100;
s.t. req_equipo_sonido: equipo_sonido >= 0.25 * televisor;
s.t. req_equipo_sonido_maxima: equipo_sonido <= televisor + 150;
s.t. prod_armado: 10 * televisor + 12 * equipo_sonido <= 3300;
s.t. pruebas: 30 * televisor + 6 * equipo_sonido <= 6000;


'''
```

```python
ampl.eval(model)
ampl.option["solver"]='highs'
ampl.solve()
```

```python
print("Resultados de la produccion: ")

cant_televisor = ampl.get_variable('televisor').value()
cant_equipo_sonido = ampl.get_variable('equipo_sonido').value()
profit = ampl.get_objective('profit').value()

print(("televisores a fabricar: "), cant_televisor)
print(("equipos de sonido a fabricar: "), cant_equipo_sonido)
print(("Beneficio: "), profit)
```
